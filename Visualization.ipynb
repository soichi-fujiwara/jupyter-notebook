##折れ線グラフ

```

import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid")

rs = np.random.RandomState(365)

# X軸の値
# cumsum(累積和)
values = rs.randn(365, 3).cumsum(axis=0)

# Y軸の値
dates = pd.date_range("1 1 2016", periods=365, freq="D")
data = pd.DataFrame(values, dates, columns=["A", "B", "C"])

#rolling(窓関数)･･･移動平均線などに利用できる
#rolling(N).nean : N毎に平均値を取り慣らす
#(N=7)であれば週次のデータとして表示する。
data = data.rolling(7).mean()

#palette:カラーマップの指定
#http://pynote.hatenablog.com/entry/matplotlib-color#%E3%82%AB%E3%83%A9%E3%83%BC%E3%83%9E%E3%83%83%E3%83%97
sns.lineplot(data=data, palette="winter", linewidth=2.5)
```

##カテゴリー変数を使用した散布図

```
import pandas as pd
import seaborn as sns

sns.set(style="whitegrid", palette="muted")

# Load the example iris dataset
#DataFrame型
iris = sns.load_dataset("iris")

#p_1:データセット
#p_2:凡例
#p_3:X軸項目
iris = pd.melt(iris, "species", var_name="measurement")

sns.swarmplot(x="measurement", y="value", hue="species",
              palette=["r", "c", "y"], data=iris)
```

##ヒストグラム付き散布図

```
import pandas as pd
import seaborn as sns
 
# レストランのチップの額のデータセットをロードします。
tips = sns.load_dataset("tips")
 
# 背景を白に設定
sns.set(style="white", color_codes=True,palette="summer")
 
# グラフを出力
sns.jointplot(x="total_bill", y="tip", data=tips)
```
